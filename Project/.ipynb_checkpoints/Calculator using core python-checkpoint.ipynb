**Simple Calculator**

```python
Defining all calculation logic inside the def soo can be called inside the calculator funstion
```

```python
def addition(x,y):
    return x + y
```

```python
def subtraction(x,y):
    return y - x
```

```python
def multiply(x,y):
    return x * y
```

```python
def divide(x,y):
    if y == 0:
        return "Error! Paglu hoo kya!! Cannot divide by Zero "
    else:
        return x / y
```

```python
def square(x):
    return x ** 2
```

```python
def root(x):
    if x < 0:
        return "Enter the valid numnber"
        
    guess = x/2.0
    tolerance = 0.000001

    while True:

        new_guess = (guess + x / guess)/2

        difference = abs(new_guess - guess)

        if difference < tolerance:
            return new_guess

        guess = new_guess
```

```python
root(25)
```
