**Simple Calculator**

```python
Defining all calculation logic inside the def soo can be called inside the calculator funstion
```

```python
def addition(x,y):
    return x + y
```

```python
def subtraction(x,y):
    return y - x
```

```python
def multiply(x,y):
    return x * y
```

```python
def divide(x,y):
    if y == 0:
        return "Error! Paglu hoo kya!! Cannot divide by Zero "
    else:
        return x / y
```

```python
def calculator():
    print("Simple Calculator")
    print("Below are Opertion Enter the number of the operation")
    print("1. Addition (+)")
    print("2. Subtraction (+)")
    print("3. Multiply (+)")
    print("4. Divition (+)")

    while True:

        choice = input("Enter the choice(1/2/3/4) or q to quite")
        
        if choice.lower() == 'q':
            print("Existing from the calculator..................................!!!!")
            break
            
        if choice in ('1','2','3','4'):
            try:
                x = float(input("Enter the first number"))
                y = float(input("Enter the Second number"))
            except ValueError:
                print("Enter the valid numeric value , Too thoda nahi pura paglu hai!!!")
                continue

            if choice == '1':
                a = addition(x,y)
                print(a)

            elif choice == '2':
                s = subtraction(x,y)
                print(s)

            elif choice == '3':
                m = multiply(x,y)
                print(m)

            elif choice == '4':
                d = divide(x,y)
                print(d)

        else:
            print("invalid input")
```

```python
calculator()
```
