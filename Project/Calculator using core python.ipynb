```python
def root(x):
    if x < 0:
        return "Enter the valid numnber"
        
    guess = x/2.0
    tolerance = 0.000001

    while True:

        new_guess = (guess + x / guess)/2

        difference = abs(new_guess - guess)

        if difference < tolerance:
            return new_guess

        guess = new_guess
```
